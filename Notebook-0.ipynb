## QUESTION 1:  Willow

Willow wants a list of book titles that have multiple authors attributed to them.  Write a query to return her desired results.

```sql
-- Code here:
SELECT title
from BooksDB.dbo.books b
where (select count(distinct authors) from BooksDB.dbo.books where title = b.title) > 1;

--select * from BooksDB.dbo.books where title = 'Anatomy for the Artist';
```

```sql
-- Code here:
SELECT distinct b.title
from BooksDB.dbo.books b
where exists (select book_id from BooksDB.dbo.book_tags bt join BooksDB.dbo.tags t on bt.tag_id = t.tag_id
where t.tag_name = 'Meditation');

--select * from BooksDB.dbo.tags where tag_name = 'Meditation';

--select * from BooksDB.dbo.book_tags where tag_id = 19682;

--select * from BooksDB.dbo.books where title like '%medidation';
```

## QUESTION 3:  Alyce

Alyce wants to create a display of books with an average rating of 4.2 or higher, but she only wants books that share both title and original title.  Write a query that displays both the original title and title for Alyce, in descending order by rating.  Make sure that none of the original titles contain NULL values either.

```sql
-- Code here:
SELECT DISTINCT b1.original_title, b1.title, b1.average_rating
from BooksDB.dbo.books b1 
where b1.average_rating >= 4.2 
and b1.original_title is not NULL
and exists (
    select 1 from BooksDB.dbo.books b2 where b1.title = b2.title and b1.original_title = b2.original_title
)
order by b1.average_rating desc;
```

## QUESTION 4:  Thisbe

Thisbe is planning to make a long-term display of popular books that she will update monthly, reflecting different time periods.  She wants to make it reader-centric by including titles that readers have averagely rated higher than the book's actual average rating. 

**Part A:**   Write a query that compares the publication time with average reader ratings.

**Part B:**  She wants the list ordered by year, then the book title

```sql
-- Code here:
select title, original_publication_year, average_rating
from BooksDB.dbo.books b1 
where
(select avg(ratings_1) from BooksDB.dbo.books b2 where b1.title = b2.title) > average_rating
order BY
original_publication_year desc, title desc;

```
